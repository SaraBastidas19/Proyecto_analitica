```python
#Importamos librerías básicas
import pandas as pd # manipulacion dataframes
import numpy as np  # matrices y vectores
import matplotlib.pyplot as plt #gráfica

```

```python
data = pd.read_csv("Accidentalidad_Municipio_de__Envigado_20250514.csv")
data.head()
```

```python
data.info()
```

```python
#Corrección del tipo de datos object a categorías
data['FECHA']=data['FECHA'].astype('category')
data['HORA']=data['HORA'].astype('category')
data['DÍA DE LA SEMANA']=data['DÍA DE LA SEMANA'].astype('category')
data['CLASE DE VEHICULO']=data['CLASE DE VEHICULO'].astype('category')
data['TIPO DE SERVICIO']=data['TIPO DE SERVICIO'].astype('category')
data['SEXO']=data['SEXO'].astype('category')
data['GRAVEDAD']=data['GRAVEDAD'].astype('category')
data['CLASE DE ACCIDENTE']=data['CLASE DE ACCIDENTE'].astype('category')
data['CAUSA']=data['CAUSA'].astype('category')
data['DIRECCIÓN']=data['DIRECCIÓN'].astype('category')
data['BARRIO']=data['BARRIO'].astype('category')
data['AREA']=data['AREA'].astype('category')
data['Coordenadas']=data['Coordenadas'].astype('category')
data['TIPO DE VICTIMA']=data['TIPO DE VICTIMA'].astype('category')


data.info()
```

```python
# Sólo dejamos el registro más reciente por cada número de radicado
data.drop_duplicates(subset=['RADICADO'], keep='last', inplace=True)
# Ver cuántos duplicados existen por RADICADO
print(f'Duplicados por RADICADO: {data.duplicated(subset=["RADICADO"]).sum()}')
```

```python
data.info()
```

```python
# Variables irrelevantes para el proceso de minería
data = data.drop('TIPO DE VICTIMA',axis=1) #Eliminamos el ID por ser irrelevante, axis=1 indica que es una columna
data = data.drop('SEXO',axis=1)
data = data.drop('CLASE DE VEHICULO',axis=1)
data = data.drop('TIPO DE SERVICIO',axis=1)
data = data.drop('DIRECCIÓN',axis=1)
data = data.drop('Coordenadas',axis=1)

data.head()
```
